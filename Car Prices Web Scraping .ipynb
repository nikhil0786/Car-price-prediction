```python
brand=[]
model=[]
year=[]
fuel=[]
transmission=[]
km_driven=[]
no_of_owners=[]
price=[]
```

```python
urls=driver.find_elements_by_xpath("//li[@class='EIR5N']/a")#collecting urls of all the cars
UR=[]
for i in urls[:500]:
    UR.append(i.get_attribute('href'))#getting the url of first 100 car
for url in UR:#loop for every car in the list
    driver.get(url)
    try:                  #exception handling for nosuchelementexception
        brands=driver.find_element_by_xpath("//div[@class='_3JPEe']/div[1]/div[1]/span[2]")#locating the brand name
        brand.append(brands.text)
        
    except NoSuchElementException   as e:
        brand.append("NaN")
        
    try:                  #exception handling for nosuchelementexception
        models=driver.find_element_by_xpath("//div[@class='_3JPEe']/div[2]/div[1]/span[2]")#locating the brand name
        model.append(models.text)
        
    except NoSuchElementException   as e:
        model.append("NaN")    
        
    try:                  #exception handling for nosuchelementexception
        years=driver.find_element_by_xpath("//div[@class='_3JPEe']/div[4]/div[1]/span[2]")#locating the brand name
        year.append(years.text)
        
    except NoSuchElementException   as e:
        year.append("NaN")
    
    try:                  #exception handling for nosuchelementexception
        fuels=driver.find_element_by_xpath("//div[@class='_3JPEe']/div[5]/div[1]/span[2]")#locating the brand name
        fuel.append(fuels.text)
        
    except NoSuchElementException   as e:
        fuel.append("NaN")
    
    try:                  #exception handling for nosuchelementexception
        trans=driver.find_element_by_xpath("//div[@class='_3JPEe']/div[6]/div[1]/span[2]")#locating the brand name
        transmission.append(trans.text)
        
    except NoSuchElementException   as e:
        transmission.append("NaN")
    
    try:                  #exception handling for nosuchelementexception
        kms_driven=driver.find_element_by_xpath("//div[@class='_3JPEe']/div[7]/div[1]/span[2]")#locating the brand name
        km_driven.append(kms_driven.text)
        
    except NoSuchElementException   as e:
        km_driven.append("NaN")
        
        
    try:                  #exception handling for nosuchelementexception
        owners=driver.find_element_by_xpath("//div[@class='_3JPEe']/div[8]/div[1]/span[2]")#locating the brand name
        no_of_owners.append(owners.text)
        
    except NoSuchElementException   as e:
        no_of_owners.append("NaN")
        
        
    try:                  #exception handling for nosuchelementexception
        prices=driver.find_element_by_xpath("//span[@class='_2xKfz']")#locating the brand name
        price.append(prices.text)
        
    except NoSuchElementException   as e:
        price.append("NaN")  
```
